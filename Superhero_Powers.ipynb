```python
# 8. Show top five records having top speed of heroes of good alignment

# df.nlargest(5, 'Speed')
# df.loc[df['Alignment']=='good'].nlargest(5, 'Speed')
df[['Name','Speed']].loc[df['Alignment']=='good'].nlargest(5, 'Speed')
# Using the .nlargest(5, 'Speed') to find the top 5 in 'Speed'
```

```python
# 9. Show 5 records of super heroes who have maximum power of good alignment

# df.loc[df['Alignment']=='good'].nlargest(5, 'Power')
df[['Name','Power']].loc[df['Alignment']=='good'].nlargest(5, 'Power')
# Using the .nlargest(5, 'Power') to find the top 5 in 'Power'
```

```python
# 10. Find out how many super heroes are there with power 100 of good alignment

# df.loc[ (df['Alignment']=='good') & (df['Power']==100) ]

super_power = df[['Name','Power']].loc[ (df['Alignment']=='good') & (df['Power']==100) ]
# Here you can use multiple df['Some Value']=='Other Value'
# Use multiple values by using & to call another record

# display result
super_power
```

```python
# 11. Shape them what you got in point 10

# shape_result = df[['Name','Power']].loc[ (df['Alignment']=='good') & (df['Power']==100) ].shape

shape_result = super_power.shape

print("\nNumber of Rows :", shape_result[0])
print("\nNumber of Columns :", shape_result[1])
print("\nDataFrame Shape :", shape_result)
```

```python
# 12. Show all records from point 10

df.loc[ (df['Alignment']=='good') & (df['Power']==100) ]
# Unlike point 10 where we use df[['Name','Power']] to show specific columns only
# use only df to show all records in the csv with every column
```

```python
# 13. Retrieve total of first five records of max power of good alignment super heroes

# df[['Name','Power']].loc[ (df['Alignment']=='good') & (df['Power']==100) ].nlargest(5, 'Power')
# Retriving the first 5 by using the .nlargest(5, 'Power')
# Even though it shows nlargest, it displays first 5 by default
output_hero = df[['Name','Total','Power']].loc[ (df['Alignment']=='good') & (df['Power']==100) ].nlargest(5, 'Power')
print(output_hero)
# stored the entire result in a variable
# I switched the 'Total' and 'Power' so it shows different values for bar plot
# Bar Plot takes the first 2 column names which are 'Name' and 'Total'
```

```python
# 14. Draw a bar plot of all super heroes who are having good alignment and max power of top five only
# take same object of point 13 , show name and total in plot with green bars

X = output_hero.iloc[:,0]
Y = output_hero.iloc[:,1]

plt.bar(X, Y, color='green')
plt.title("Super Heroes")
plt.xticks(rotation = 45)   
# rotate 45 to get clearer x axis
plt.xlabel("Name")
plt.ylabel("Total")
plt.show()
```

```python
# 20. Draw a histogram for speed of super heroes having fig size 10,5 , 
# provide speed in histogram for only good alignment super heroes ,
# title should be "distribution of speed" , xlabel should be "speed"

#speed" , xlabel should be "speed"
output = ps.sqldf("select speed from df where Alignment = 'good'")

# print(output)
# Creating histogram change the size of the figure
fig, ax = plt.subplots(figsize=(10, 5))
ax.hist(output)

# Set title
ax.set_title("Distribution of speed")

# adding labels
ax.set_xlabel('Speed')
```

```python
# 21. Draw a histogram for combat of super villains having fig size 10,5 , 
# provide combat in histogram for only bad alignment super heroes ,
# title should be "distribution of combat" , xlabel should be "combat"

#combat" , xlabel should be "combat"
output = ps.sqldf("select combat from df where Alignment = 'bad'")

# print(output)
# Creating histogram change the size of the figure
fig, ax = plt.subplots(figsize=(10, 5))
ax.hist(output)

# Set title
ax.set_title("Distribution of combat")

# adding labels
ax.set_xlabel('Combat')
```
